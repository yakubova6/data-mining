## Упражнение 1.1.2
*Напишите модуль для нахождения НОД числа 8 и заданного 10-значного числа, не используя арифметические операции, но с использованием функций пользователя.*

```python
# Функция для получения правильного 10-значного числа
def get_valid_number():
    # Цикл для ввода числа
    while True:
        # Получаем ввод от пользователя
        num_str = input("Введите 10-значное число: ")
        
        # Проверяем: длина=10, все символы являются цифрами, число не начинается с 0
        if len(num_str) == 10 and num_str.isdigit() and not num_str.startswith('0'):
            # Преобразуем строку в число и возвращаем его
            return int(num_str)
        
        # Сообщение об ошибке при неправильном вводе
        print("Ошибка! Число должно быть 10-значным и не начинаться с нуля")

# Функция для нахождения НОД числа 8 и 10-значного числа
def gcd_with_8(x):
    # Преобразуем число в двоичную строку без префикса '0b'
    binary_str = bin(x)[2:]
    
    # Проверяем последние три бита (делимость на 8)
    if len(binary_str) >= 3 and binary_str[-3:] == '000':
        return 8
    
    # Проверяем последние два бита (делимость на 4)
    elif len(binary_str) >= 2 and binary_str[-2:] == '00':
        return 4
    
    # Проверяем последний бит (делимость на 2)
    elif len(binary_str) >= 1 and binary_str[-1] == '0':
        return 2
    
    # Если число нечетное - НОД равен 1
    else:
        return 1

# Основная программа
print("---------------------------------------")
# Получаем число от пользователя
number = get_valid_number()
# Вычисляем НОД
result = gcd_with_8(number)
# Выводим результат
print(f"НОД числа 8 и {number} равен {result}")
print("---------------------------------------")
```

## Упражнение 1.2.2

*Создайте список из 100 первых элементов последовательности частичных сумм ряда Фибоначчи. Выведите число значащих цифр у элемента последовательности с заданным номером.*

```python
# Функция для генерации частичных сумм ряда Фибоначчи
def generate_fibonacci_sums(n):
    fibonacci = [0, 1]  # Начальные числа Фибоначчи
    partial_sums = [0, 1]  # Начальные частичные суммы
    
    # Генерируем остальные числа и суммы
    for i in range(2, n):
        # Вычисляем следующее число Фибоначчи
        next_fib = fibonacci[i-1] + fibonacci[i-2]
        fibonacci.append(next_fib)
        
        # Вычисляем следующую частичную сумму
        next_sum = partial_sums[i-1] + next_fib
        partial_sums.append(next_sum)
    
    return partial_sums

# Функция для подсчета значащих цифр в числе
def count_significant_digits(number):
    # Преобразуем число в строку и убираем ведущие нули
    num_str = str(number).lstrip('0')
    
    # Если после удаления нулей строка пустая, значит число было 0
    if not num_str:
        return 1  # Число 0 имеет одну значащую цифру
    
    return len(num_str)

# Основная программа
print("---------------------------------------")
print("Создание списка из 100 частичных сумм ряда Фибоначчи")
print("---------------------------------------")

# Генерируем список из 100 частичных сумм
fibonacci_sums = generate_fibonacci_sums(100)

# Выводим первые 10 сумм для демонстрации
print("Первые 10 частичных сумм:")
for i in range(10):
    print(f"S{i} = {fibonacci_sums[i]}")

print("...")
print(f"S99 = {fibonacci_sums[99]}")

print("---------------------------------------")
print("Подсчет значащих цифр для элемента с заданным номером")
print("---------------------------------------")

# Запрашиваем у пользователя номер элемента
while True:
    try:
        index = int(input("Введите номер элемента (0-99): "))
        if 0 <= index <= 99:
            break
        else:
            print("Номер должен быть в диапазоне от 0 до 99.")
    except ValueError:
        print("Пожалуйста, введите целое число.")

# Получаем элемент и подсчитываем значащие цифры
element = fibonacci_sums[index]
significant_digits = count_significant_digits(element)

# Выводим результат
print(f"Элемент с номером {index}: {element}")
print(f"Количество значащих цифр: {significant_digits}")
print("---------------------------------------")
```

## Упражнение 1.3.1
*Создайте короткий скрипт, находящий все нули заданной алгебраической функции на заданном интервале (переменная формата float).*

```python
# Функция для поиска нулей функции методом дихотомии
from math import tan

def find_zeros(f, a, b, epsilon=1e-6, max_iterations=100):
    zeros = []  # Список для найденных нулей
    
    # Проверяем, является ли левая граница интервала (a) нулем функции
    # Если |f(a)| очень близко к 0 (в пределах точности epsilon), то добавляем точку в список найденных нулей
    if abs(f(a)) < epsilon:
        zeros.append(a)
        
    # Проверяем, является ли правая граница интервала (b) нулем функции    
    if abs(f(b)) < epsilon:
        zeros.append(b)
    
    stack = [(a, b)] # Стек для хранения интервалов для исследования
    
    # Пока стек не пуст И количество найденных нулей меньше максимального
    while stack and len(zeros) < max_iterations:
        # Извлекаем последний добавленный интервал из стека
        a, b = stack.pop() 
        # Вычисляем середину текущего интервала
        c = (a + b) / 2.0 
        
        # Проверяем, является ли середина интервала (c) нулем функции
        # Если |f(с)| очень близко к 0 (в пределах точности epsilon), то добавляем точку в список найденных нулей
        if abs(f(c)) < epsilon:
            zeros.append(c)
            continue
        
        # Проверяем левую половину [a, c]
        # Если знак меняется - есть ноль
        if f(a) * f(c) < 0:  
            # Если интервал достаточно мал, то мы достигли нужной точности
            if (b - a) < epsilon:  
                zeros.append(c)
            # Иначе добавляем интервал для дальнейшего исследования
            else:
                stack.append((a, c))  
        
        # Проверяем правую половину [c, b]
        # Если знак меняется - есть ноль
        if f(c) * f(b) < 0:  
            # Если интервал достаточно мал, то мы достигли нужной точности
            if (b - a) < epsilon: 
                zeros.append(c)
            # Иначе добавляем интервал для дальнейшего исследования
            else:
                stack.append((c, b)) 
    
    return zeros

# Пример функции: f(x) = x² - 4
def example_function(x):
    return tan(x)

# Основная программа
if __name__ == "__main__":
    print("---------------------------------------")
    print("Поиск нулей функции методом дихотомии")
    print("---------------------------------------")
    
    # Задаем интервал поиска и точность
    a = -1.0
    b = 9.0
    epsilon = 1e-6  # Точность поиска
    
    # Ищем нули функции
    zeros = find_zeros(example_function, a, b, epsilon)
    
    # Выводим результаты
    print(f"Функция: f(x) = x² - 4")
    print(f"Интервал: [{a}, {b}]")
    print(f"Точность: {epsilon}")
    print(f"Найдено нулей: {len(zeros)}")
    
    # Выводим найденные нули
    for i, zero in enumerate(zeros, 1):
        print(f"Ноль #{i}: x = {zero:.6f}, f(x) = {example_function(zero):.2e}")
    
    print("---------------------------------------")
```

## Упражнение 1.3.3
*Скачайте любую статью (например, с Habr.ru) и сформируйте из нее строку. Посчитайте частотности для всех слов, содержащихся в ней.*

```python
import requests  # Для отправки HTTP-запросов и получения веб-страниц
from bs4 import BeautifulSoup  # Для парсинга HTML и извлечения данных из веб-страниц
import re  # Для работы с регулярными выражениями (поиск и фильтрация текста)
from collections import Counter  # Для подсчета частотности слов
import nltk  # Natural Language Toolkit - библиотека для обработки естественного языка
from nltk.tokenize import word_tokenize  # Функция для разбиения текста на отдельные слова (токены)
import pymorphy3  # Морфологический анализатор для русского языка (для приведения слов к нормальной форме)

# Загрузка ресурсов nltk для токенизации (разбиения на слова)
try:
    # Проверяем, загружен ли уже ресурс для токенизации
    nltk.data.find('tokenizers/punkt')
except LookupError:
    # Если ресурс не найден, загружаем его
    nltk.download('punkt')

# URL статьи для анализа на Habr
url = "https://habr.com/ru/articles/945452/"

# Отправка GET-запроса к указанному URL и получение HTML-страницы
print("Отправка запроса к Habr...")
response = requests.get(url)  # Отправляем HTTP GET запрос
response.raise_for_status()  # Проверяем, что запрос выполнен успешно (статус 200), иначе вызываем исключение

# Парсинг HTML-кода страницы с помощью BeautifulSoup
print("Парсинг HTML...")
soup = BeautifulSoup(response.text, 'html.parser')  # Создаем объект BeautifulSoup для работы с HTML

# Поиск основного контента статьи на странице
# Сначала ищем тег <article>, который содержит основную статью
article = soup.find('article')

# Проверяем, найден ли контент статьи
if article:
    # Извлечение текста из всех параграфов (тегов <p>) внутри найденного контейнера статьи
    paragraphs = article.find_all('p')  # Находим все теги <p> в статье
    text = ' '.join([p.get_text() for p in paragraphs])  # Объединяем текст всех параграфов в одну строку
    print(f"Текст статьи получен ({len(text)} символов)")  # Выводим информацию о длине текста
else:
    # Если статья не найдена, вызываем исключение с сообщением об ошибке
    raise Exception("Статья не найдена на странице")

# Токенизация текста - разбиение текста на отдельные слова (токены)
print("Токенизация текста...")
tokens = word_tokenize(text, language='russian')  # Используем NLTK для разбиения текста на слова
print(f"Получено {len(tokens)} токенов")  # Выводим количество полученных токенов

# Инициализация морфологического анализатора pymorphy3 для русского языка
print("Инициализация морфологического анализатора...")
morph = pymorphy3.MorphAnalyzer()  # Создаем объект морфологического анализатора

# Создаем пустой список для хранения лемм (слов в нормальной форме)
lemmas = []
# Счетчик для отслеживания количества обработанных слов
processed_words = 0

# Обработка каждого токена (слова) для приведения к нормальной форме
print("Приведение слов к нормальной форме...")

# Проходим по всем токенам с их индексами
for i, token in enumerate(tokens):
    # Приведение токена к нижнему регистру
    token_lower = token.lower()
    
    # Фильтрация токенов: от начала до конца строки только русские буквы и дефис
    if not re.match(r'^[а-яё-]+$', token_lower):
        continue  # Пропускаем слвово если НЕТ совпадения
    
    try:
        # Морфологический анализ слова
        parsed = morph.parse(token_lower)[0]  # Берем наиболее вероятный разбор
        lemma = parsed.normal_form  # Извлекаем нормальную форму
        
        # Добавляем лемму (слово в нормальной форме) в список
        lemmas.append(lemma)
        processed_words += 1  # Увеличиваем счетчик обработанных слов
        
    except Exception as e:
        # В случае ошибки используем оригинальное слово
        lemmas.append(token_lower)

# Выводим информацию о количестве обработанных слов
print(f"Обработано {processed_words} слов")

# Подсчет частотности слов
print("Подсчет частотности...")
word_freq = Counter(lemmas)

# Вывод 20 самых частых слов
print("\n20 самых частых слов в статье (после лемматизации):")
print("-" * 50)
for word, count in word_freq.most_common(20):
    print(f"{word}: {count}")

# Дополнительная статистика
print(f"\nСтатья на Habr: {url}")
print(f"Общее количество обработанных слов: {len(lemmas)}") 
print(f"Уникальные слова (после лемматизации): {len(word_freq)}")
```

## Упражнение 1.4.3
*Создайте класс Frac, типичный экземпляр которого является обыкновенной дробью. Опишите методы обращения, сложения и умножения.*

```python
# Импортируем функцию для нахождения наибольшего общего делителя
from math import gcd

# Класс для работы с обыкновенными дробями
class Frac:
    # Конструктор класса - вызывается при создании нового объекта
    def __init__(self, numerator, denominator):
        # Проверяем, что знаменатель не равен нулю
        if denominator == 0:
            raise ValueError("Знаменатель не может быть нулем")
        
        # Находим наибольший общий делитель для сокращения дроби
        gcd_value = gcd(abs(numerator), abs(denominator))
        
        # Сокращаем дробь, деля числитель и знаменатель на НОД
        self.numerator = numerator // gcd_value
        self.denominator = denominator // gcd_value
        
        # Нормализуем знаки - делаем знаменатель всегда положительным
        if self.denominator < 0:
            self.numerator = -self.numerator
            self.denominator = -self.denominator

    # Магический метод для строкового представления объекта
    def __repr__(self):
        # Возвращаем строку в формате "числитель/знаменатель"
        return f"{self.numerator}/{self.denominator}"

    # Метод для создания обратной дроби
    def inverse(self):
        # Проверяем, что числитель не равен нулю
        if self.numerator == 0:
            raise ZeroDivisionError("Нельзя обратить дробь с нулевым числителем")
        
        # Создаем новую дробь, меняя местами числитель и знаменатель
        return Frac(self.denominator, self.numerator)

    # Магический метод для сложения дробей (оператор +)
    def __add__(self, other_fraction):
        # Вычисляем новый числитель
        new_numerator = self.numerator * other_fraction.denominator + other_fraction.numerator * self.denominator
        # Вычисляем новый знаменатель
        new_denominator = self.denominator * other_fraction.denominator
        # Создаем и возвращаем новую дробь
        return Frac(new_numerator, new_denominator)

    # Магический метод для умножения дробей (оператор *)
    def __mul__(self, other_fraction):
        # Вычисляем новый числитель
        new_numerator = self.numerator * other_fraction.numerator
        # Вычисляем новый знаменатель
        new_denominator = self.denominator * other_fraction.denominator
        # Создаем и возвращаем новую дробь
        return Frac(new_numerator, new_denominator)


# Демонстрация работы класса
if __name__ == "__main__":
    print("=" * 55)
    print("Демонстрация выполнения различных методов класса Frac")
    print("=" * 55)
    
    # Создаем две дроби
    fraction1 = Frac(6, 8)  
    fraction2 = Frac(2, 6)  
    
    # Выводим дроби и операции с ними
    print(f"Дробь 1: {fraction1}")
    print(f"Дробь 2: {fraction2}")
    print(f"Обратная дробь к Дроби 1: {fraction1.inverse()}")
    print(f"Сумма дробей: {fraction1} + {fraction2} = {fraction1 + fraction2}")
    print(f"Произведение дробей: {fraction1} * {fraction2} = {fraction1 * fraction2}")
```

## Упражнение 1.5.1
*Постройте графики стандартного отклонения температур по месяцам года для Москвы и Анадыря, пользуясь данными rp5.ru.*

```python
# Импортируем необходимые библиотеки
import pandas as pd  # для работы с табличными данными
import numpy as np   # для математических вычислений
import matplotlib.pyplot as plt  # для построения графиков

# Загружаем данные из CSV файлов
moscow_data = pd.read_csv('moscow.csv', sep=';', skiprows=6, encoding='utf-8')
anadyr_data = pd.read_csv('anadyr.csv', sep=';', skiprows=6, encoding='utf-8')

# Функция для обработки метеорологических данных
def process_weather_data(weather_table, city_name): 
    # Переименовываем столбцы для удобства
    # Первый столбец - дата и время, второй - температура
    weather_table = weather_table.rename(columns={
        weather_table.columns[0]: 'datetime', 
        weather_table.columns[1]: 'temperature'
    })
    
    # Преобразуем текстовую дату в формат datetime: день.месяц.год час:минута
    weather_table['datetime'] = pd.to_datetime(weather_table['datetime'], format='%d.%m.%Y %H:%M')
    
    # Преобразуем температуру из текста в числа
    weather_table['temperature'] = pd.to_numeric(weather_table['temperature'])
    
    # Извлекаем номер месяца из даты (1-12)
    weather_table['month'] = weather_table['datetime'].dt.month
    # Извлекаем число месяца из даты (1-31)
    weather_table['day'] = weather_table['datetime'].dt.day
    
    # Группируем данные по месяцу и дню, вычисляем среднесуточную температуру
    daily_average_temperatures = weather_table.groupby(['month', 'day'])['temperature'].mean().reset_index()

    # Вычисляем стандартное отклонение температур для каждого месяца
    monthly_standard_deviation = daily_average_temperatures.groupby('month')['temperature'].std()
    
    return monthly_standard_deviation

# Обрабатываем данные для обоих городов
moscow_standard_deviation = process_weather_data(moscow_data, 'Москва')
anadyr_standard_deviation = process_weather_data(anadyr_data, 'Анадырь')

# Подготавливаем данные для построения графика
month_numbers = range(1, 13)  # номера месяцев от 1 до 12
month_abbreviations = ['Янв', 'Фев', 'Мар', 'Апр', 'Май', 'Июн', 
                       'Июл', 'Авг', 'Сен', 'Окт', 'Ноя', 'Дек']  # сокращенные названия месяцев

# Создаем новое окно для графика размером 12x6 дюймов
plt.figure(figsize=(12, 6))

# Строим линию для Москвы с кружками в точках данных
plt.plot(
    month_numbers, 
    moscow_standard_deviation.values, 
    'o-',  # стиль: кружки с соединением линиями
    label='Москва', 
    linewidth=2.5,  # толщина линии
    markersize=6,   # размер маркеров
    color='#2E86AB'     # цвет линии
)

# Строим линию для Анадыря с квадратиками в точках данных
plt.plot(
    month_numbers, 
    anadyr_standard_deviation.values, 
    's-',  # стиль: квадратики с соединением линиями
    label='Анадырь', 
    linewidth=2.5,  # толщина линии
    markersize=6,   # размер маркеров
    color='#A23B72'    # цвет линии
)

# Добавляем подписи значений над точками для Москвы
for month_index, deviation_value in enumerate(moscow_standard_deviation.values):
    plt.text(
        month_index + 1,  # позиция по X (номер месяца)
        deviation_value + 0.2,  # позиция по Y (немного выше точки)
        f'{deviation_value:.1f}',  # текст: значение с одним знаком после запятой
        ha='center',      # горизонтальное выравнивание по центру
        va='bottom',      # вертикальное выравнивание по низу
        fontsize=9,       # размер шрифта
        fontweight='bold', # жирный шрифт
        color='#2E86AB'       # цвет текста
    )

# Добавляем подписи значений над точками для Анадыря
for month_index, deviation_value in enumerate(anadyr_standard_deviation.values):
    plt.text(
        month_index + 1,  # позиция по X (номер месяца)
        deviation_value + 0.2,  # позиция по Y (немного выше точки)
        f'{deviation_value:.1f}',  # текст: значение с одним знаком после запятой
        ha='center',      # горизонтальное выравнивание по центру
        va='bottom',      # вертикальное выравнивание по низу
        fontsize=9,       # размер шрифта
        fontweight='bold', # жирный шрифт
        color='#A23B72'      # цвет текста
    )

# Добавляем заголовок графика
plt.title(
    'Стандартное отклонение температур по месяцам 2024 года', 
    fontsize=14,      # размер шрифта
    fontweight='bold' # жирный шрифт
)

# Добавляем подпись оси X
plt.xlabel('Месяц', fontsize=12)

# Добавляем подпись оси Y
plt.ylabel('Стандартное отклонение температуры (°C)', fontsize=12)

# Настраиваем метки на оси X (номера месяцев и их названия)
plt.xticks(month_numbers, month_abbreviations)

# Добавляем легенду (обозначение линий)
plt.legend(fontsize=11)

# Добавляем сетку на график
plt.grid(True, alpha=0.3)  # alpha=0.3 - прозрачность сетки

# Показываем график
plt.show()

# Выводим статистику в консоль
print("=" * 60)
print("ИТОГОВАЯ СТАТИСТИКА СТАНДАРТНЫХ ОТКЛОНЕНИЙ")
print("=" * 60)

# Вычисляем средние значения стандартных отклонений
moscow_average_deviation = moscow_standard_deviation.mean()
anadyr_average_deviation = anadyr_standard_deviation.mean()

# Выводим статистику для Москвы
print("\nМосква:")
print(f"- Среднее стандартное отклонение: {moscow_average_deviation:.2f}°C")
print(f"- Максимальное: {moscow_standard_deviation.max():.2f}°C ({month_abbreviations[moscow_standard_deviation.argmax()]})")
print(f"- Минимальное: {moscow_standard_deviation.min():.2f}°C")

# Выводим статистику для Анадыря
print("\nАнадырь:")
print(f"- Среднее стандартное отклонение: {anadyr_average_deviation:.2f}°C")
print(f"- Максимальное: {anadyr_standard_deviation.max():.2f}°C ({month_abbreviations[anadyr_standard_deviation.argmax()]})")
print(f"- Минимальное: {anadyr_standard_deviation.min():.2f}°C")

# Сравниваем результаты двух городов
print(f"\nСравнение:")
difference = abs(moscow_average_deviation - anadyr_average_deviation)
if moscow_average_deviation > anadyr_average_deviation:
    print(f"Москва имеет большую изменчивость температур на {difference:.2f}°C")
else:
    print(f"Анадырь имеет большую изменчивость температур на {difference:.2f}°C")
```

## Упражнение 1.5.2
*Постройте три столбчатые диаграммы населения 5 стран мира (по вашему выбору) в 2000, 2010 и 2020 годах по данным ресурса worldometer.*

```python
import requests  # Для выполнения HTTP-запросов к сайту
from bs4 import BeautifulSoup  # Для парсинга HTML-контента
import matplotlib.pyplot as plt  # Для построения графиков
import time  # Для добавления задержек между запросами

# Функция для получения данных о населении страны за указанные годы с Worldometer
def get_country_population(country_url, years):
    
    try:
        # Добавляем задержку в 1 секунду между запросами
        time.sleep(1)
        # Выполняем GET-запрос к указанному URL с таймаутом 15 секунд
        response = requests.get(country_url, timeout=15)
        # Проверяем успешность запроса (код статуса 200)
        response.raise_for_status()
        
        # Создаем объект BeautifulSoup для парсинга HTML-контента
        soup = BeautifulSoup(response.content, 'html.parser')
        # Ищем таблицу с указанным классом
        table = soup.find('table', class_='datatable w-full border border-zinc-200')
        # Если таблица не найдена, возвращаем None
        if not table:
            return None
        
        # Создаем словарь для хранения данных о населении
        population_data = {}
        # Получаем все строки таблицы, пропуская заголовок (первую строку)
        rows = table.find_all('tr')[1:]
        
        # Проходим по каждой строке таблицы
        for row in rows:
            # Получаем все ячейки в текущей строке
            cols = row.find_all('td')
            # Проверяем, что в строке достаточно столбцов (хотя бы 2)
            if len(cols) >= 2:
                # Извлекаем текст из первого столбца (год) и удаляем лишние пробелы
                year_text = cols[0].text.strip()
                # Извлекаем текст из второго столбца (население), удаляем запятые и пробелы
                population_text = cols[1].text.strip().replace(',', '').replace(' ', '')
                
                # Проверяем, что текст года состоит только из цифр
                if year_text.isdigit():
                    # Преобразуем год в целое число
                    year = int(year_text)
                    # Проверяем, что текст населения состоит только из цифр
                    if population_text.isdigit():
                        # Преобразуем население в целое число
                        population = int(population_text)
                        # Если год находится в списке запрошенных лет, сохраняем данные
                        if year in years:
                            population_data[year] = population
        
        # Возвращаем словарь с данными о населении
        return population_data
        
    except Exception as e:
        # В случае ошибки выводим сообщение и возвращаем None
        print(f"Ошибка: {e}")
        return None

# Функция для построения столбчатых диаграмм для выбранных стран 
def plot_population_charts():
    # Словарь с URL-адресами страниц стран на Worldometer
    countries = {
        'Россия': 'https://www.worldometers.info/world-population/russia-population/',
        'США': 'https://www.worldometers.info/world-population/us-population/',
        'Германия': 'https://www.worldometers.info/world-population/germany-population/',
        'Турция': 'https://www.worldometers.info/world-population/turkey-population/',
        'Пакистан': 'https://www.worldometers.info/world-population/pakistan-population/'
    }
    
    # Список лет, за которые нужно получить данные
    years = [2000, 2010, 2020]
    # Словарь для хранения данных по всем странам
    all_data = {}
    
    # Проходим по всем странам в словаре
    for country, url in countries.items():
        # Выводим сообщение о начале сбора данных для текущей страны
        print(f"Получаем данные для {country}...")
        # Получаем данные о населении для текущей страны
        data = get_country_population(url, years)
        # Если данные получены успешно, сохраняем их
        if data:
            all_data[country] = data
            # Выводим полученные данные для отладки
            print(f"Данные для {country}: {data}")
    
    # Создаем фигуру с тремя подграфиками (1 строка, 3 столбца)
    fig, axes = plt.subplots(1, 3, figsize=(20, 7))
    # Список цветов для столбцов разных стран
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']
    
    # Проходим по каждому году (2000, 2010, 2020)
    for i, year in enumerate(years):
        # Списки для хранения названий стран и их населения за текущий год
        countries_list = []
        populations = []
        
        # Проходим по всем странам и их данным
        for country, data in all_data.items():
            # Если для текущего года есть данные, добавляем их в списки
            if year in data:
                countries_list.append(country)
                populations.append(data[year])
        
        # Если есть данные для текущего года, строим диаграмму
        if countries_list:
            # Создаем столбчатую диаграмму
            bars = axes[i].bar(countries_list, populations, 
                              color=colors[:len(countries_list)], 
                              alpha=0.8, edgecolor='black', linewidth=1)
            
            # Устанавливаем заголовок для текущего подграфика
            axes[i].set_title(f'Население в {year} году', fontsize=14, fontweight='bold')
            # Устанавливаем подпись для оси X
            axes[i].set_xlabel('Страны', fontsize=12)
            # Устанавливаем подпись для оси Y
            axes[i].set_ylabel('Население, млн человек', fontsize=12)
            # Поворачиваем подписи на оси X на 45 градусов
            axes[i].tick_params(axis='x', rotation=45)
            # Добавляем сетку на оси Y
            axes[i].grid(axis='y', alpha=0.3, linestyle='--')
            
            # Отключаем научную нотацию на оси Y
            axes[i].ticklabel_format(style='plain', axis='y')
            # Форматируем значения на оси Y (показываем в миллионах с одной десятичной цифрой)
            axes[i].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1000000:.1f}'))
            
            # Добавляем подписи значений на каждый столбец
            for bar in bars:
                # Получаем высоту текущего столбца
                height = bar.get_height()
                # Форматируем подпись: для больших чисел показываем целое значение, для малых - с десятичной точкой
                label = f'{height/1000000:.1f}M'  # Всегда показываем с одной десятичной цифрой
                
                # Добавляем текст над столбцом
                axes[i].text(bar.get_x() + bar.get_width()/2., height * 1.01,
                            label, ha='center', va='bottom', 
                            fontweight='bold', fontsize=10)
            
            # Устанавливаем пределы оси Y (сверху добавляем 15% места для подписей)
            max_pop = max(populations)
            axes[i].set_ylim(0, max_pop * 1.15)
    
    # Добавляем общий заголовок для всех графиков
    plt.suptitle('Сравнение населения стран по данным Worldometer', 
                 fontsize=16, fontweight='bold', y=0.98)
    # Автоматически настраивает отступы и расположение элементов графика
    plt.tight_layout()
    # Показываем графики
    plt.show()
    
    # Выводим данные в текстовом виде
    print("\n" + "="*60)
    print("ДАННЫЕ О НАСЕЛЕНИИ СТРАН")
    print("="*60)
    # Заголовок таблицы
    print(f"{'Страна':<15} {'2000 год':<15} {'2010 год':<15} {'2020 год':<15}")
    print("="*60)
    
    # Проходим по всем странам и выводим их данные
    for country, data in all_data.items():
        # Начинаем строку с названия страны
        line = f"{country:<15}"
        # Добавляем данные для каждого года
        for year in years:
            if year in data:
                # Форматируем число с разделителями тысяч
                pop_str = f"{data[year]:,}".replace(",", " ")
                line += f"{pop_str:<15}"
            else:
                # Если данных нет, показываем "N/A"
                line += f"{'N/A':<15}"
        print(line)
    
    print("="*60)

# Основная программа
if __name__ == "__main__":
    # Выводим сообщение о начале работы программы
    print("Сбор данных о населении стран...")
    # Вызываем функцию построения графиков
    plot_population_charts()
```
